```python
import pandas as pd 
import seaborn as sns 
sns.set(color_codes=True)
```

```python
dataset = pd.read_csv ("adm_data.CSV1.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
sns.barplot(dataset['GRE Score'], dataset['CGPA'])
```

```python
sns.displot(dataset['GRE Score'])
```

```python
sns.pairplot(dataset[['GRE Score', 'CGPA', 'Research', 'University Rating', 'SOP']])
```

```python
sns.stripplot(dataset['SOP'], dataset['GRE Score' ], jitter =True )
```

```python
sns.boxplot(dataset['GRE Score'], dataset['CGPA'], hue=dataset['Research'])
```

```python
sns.pointplot(dataset['GRE Score'], dataset['CGPA'], hue=dataset['Research'])
```

```python
sns.lmplot(x='GRE Score', y='CGPA', hue='Research', data=dataset)
```

```python
sns.distplot(dataset['GRE Score'])
```

```python
sns.jointplot(dataset['GRE Score'], dataset['CGPA'], hue=dataset['Research'])
```
